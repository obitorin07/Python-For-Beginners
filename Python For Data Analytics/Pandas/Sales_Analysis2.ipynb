```python
#Importing Libraries and Loading Dataset
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/sales.csv")
```

```python
#Displaying Dataset Head
df.head()
```

```python
#lets rename column

df.rename(columns = {"TotalSales":"total_cost"},inplace = True)
```

```python
#Summary Statistics
df.describe()
```

```python
#Total Sales
total_sales = df.total_cost.sum()
print("The total sales:",total_sales)
```

```python
#Average Sales

avg_sales = df.total_cost.mean()
print("The average sales:",avg_sales)
```

```python
#Total Sales by Category
total_sales_category = df.groupby("Category").total_cost.sum()
print("Total sales by Category:",total_sales_category)
```

```python
#Average Sales by Category

avg_sales_cat = df.groupby("Category").total_cost.mean()
print("Average sales by category:",avg_sales_cat)
```

```python
#Visualizations: Total Sales by Category
plt.figure(figsize=(10,6))
total_sales_category.plot(kind ="bar",xlabel="Category",ylabel="Sales")
```

```python
#Visualizations: Average Sales by Category

avg_sales_cat.plot(kind ="bar", xlabel="Category", ylabel="Avg Sales", title ="Average Sales", color ="skyblue")
```
