#Data Analysis with Python: Customer Purchase Patterns | @RayVision

```python
import pandas as pd
```

```python
# Load the dataset
df =pd.read_csv("/content/customer_purchase.csv")
```

```python
df.head(10)
```

```python
# Total Purchases by Customer
total_purchased = df.groupby("Customer Name").Sales.sum()
print("Total Purchase By Customer:",total_purchased)
```

```python
# Average Purchase Value by Customer
avg_sales_pur = df.groupby("Customer Name").Sales.mean()
print("Average Sales By Customer",avg_sales_pur.round())
```

```python
# Customer with the Highest Total Purchases
highest_total_cus = total_purchased.idxmax()
print(f"Customer with the highest total purchases:{highest_total_cus} with {total_purchased[highest_total_cus]}" )
```

```python
# Top 5 Customers by Total Purchases
top5_cust = total_purchased.nlargest(5)
print("Top 5 customers",top5_cust)
```

```python
# Analyzing Purchase Trends
analysing = df.groupby(["Customer Name","Product"]).Sales.sum().unstack().fillna(0) # it is like pivot table actually it is pivot table
analysing
```

```python
# User Input - Specific Customer Analysis
customer_name = input("Enter the customer name to find their total and average purchase: ")
customer_data = df[df['Customer Name'] == customer_name]
total_purchase = customer_data['Sales'].sum()
average_purchase = customer_data['Sales'].mean()

print(f"\nTotal Purchase for {customer_name}: {total_purchase}")
print(f"Average Purchase for {customer_name}: {average_purchase}")
```
