#[^A-Za-z] and ^[A-Za-z]

1.  [^A-Za-z]
* here inside ^ this work as not



---


2.  ^[A-Za-z]
* here it work as starting value


```python
password = "donttell@123$##anyone"

pattern1 = r'[^A-za-z]' # not operator
pattern2 = r'^[A-za-z]' #starting value
replace1 = re.sub(pattern1,'~',password)
replace2 = re.sub(pattern2,'---',password)

print("__________________________")

print(f"here using inside '^' {replace1}")
print("\n\n")

print(f"here using outside '^' {replace2}")
```
