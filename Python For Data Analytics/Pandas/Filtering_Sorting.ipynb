```python
import pandas as pd
# Load dataset
df = pd.read_csv("/content/sales.csv")
```

```python
df.head()
```

```python
#rename column
df.rename(columns={"TotalSales":"sales"},inplace=True)
```


#### Method 1: Filtering Data by Condition




```python
# Method 1: Filtering Data by Condition
filter = df[df.sales >150]
filter.head(10)
```

#### Method 2: Sorting Data by a Column



```python
# Method 2: Sorting Data by a Column
filter_col = df.sort_values(by = "sales", ascending=True) # it returns data in desc order 9 8 7 like this
filter_col
```


#### Method 3: Filtering and Sorting Combined



```python
# Method 3: Filtering and Sorting Combined
last_step =filter.sort_values(by ="sales" , ascending=False) #filtered data
last_step
```
