```python
import pandas as pd
```

```python
# Load dataset
df =pd.read_csv("/content/sales_data_with_nulls.csv")
```

```python
# Display dataset head
df.head()
```

```python
# Handling missing values by filling with mean
df.isna().sum()
```

```python
df.fillna(df.Sales.mean(), inplace =True)
```

```python
# Grouping data by 'customer' and aggregating 'sales'
group_Data_cust = df.groupby("Customer")['Sales'].agg(['sum','count','mean'])
group_Data_cust
```

```python
#doing same thing using like alias

groupby_cust = df.groupby("Customer").agg(total_sales =("Sales","sum"),avg_sales =("Sales",'mean'),count_sales=('Sales','count'))
groupby_cust
```

```python
# Grouping data by 'product' and aggregating 'sales'
groupby_pro = df.groupby('Product')['Sales'].agg(['sum','mean','count','median']) #median is middle value
groupby_pro
```
