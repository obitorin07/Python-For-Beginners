#Filtering and Selecting Data with Pandas

* Selecting Specific Columns
* Filtering Rows Based on Condition
* Label-based Selection with loc
* Position-based Selection with iloc

```python
import pandas as pd
```

```python
#load dataset
df = pd.read_csv("/content/marks.csv")
```

```python
#display top 5
df.head()
```

```python
df.shape # 49 rows and 3 col
```

```python
# Selecting specific columns
name_age =df[['Name','Age']]
name_age
```

```python
# Filtering rows based on condition
df_age_25 = df[df.Age > 25]
df_age_25.shape
```

```python
df_score_85 = df[df['Score']>85]
df_score_85.shape
```

```python
# Using loc for label-based selection
df_loc_marks = df.loc[df['Score'] > 90]
df_loc_marks
```

```python
# Using iloc for position-based selection

df.iloc[0:5,0:]
# display top 5 data
```
