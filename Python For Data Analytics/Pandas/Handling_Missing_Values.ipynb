```python
# Method 1: Identifying Missing Values
```

```python
finding_na = df.isna().sum()
finding_na
```

```python
# Method 2: Filling Missing Values
```

```python
#lets fill with average sales

avg_data = df.Sales.mean()
# df.fillna(avg_data, inplace= True) #now it will replace

#lets use another method

df.Sales.replace(np.NaN , avg_data , inplace =True)
```

```python
# Method 3: Dropping Missing Values
```

```python
df.dropna(inplace =True)
```
