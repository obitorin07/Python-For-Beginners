#Data Analysis with Python: Merging and Joining DataFrames with Pandas

```python
import pandas as pd
```

```python
# Load the datasets
df1 = pd.read_csv("/content/emp_details.csv")
df2 = pd.read_csv("/content/emp_salaries.csv")
```

```python
df1.tail()
```

```python
# Set index to 'ID'
df1.set_index('ID')
df2.set_index("ID")
```

```python
# Inner Join
inner_join = pd.merge(df1 , df2 ,on ='ID', how ='inner')
inner_join # it takes only common values from both tables
```

```python
# Outer Join
outer_join = pd.merge(df1 , df2, on ="ID", how="outer")
outer_join # it takes all values including non common values also
```

```python
# Left Join
left_join =  pd.merge(df1, df2, on='ID',how = 'left')
left_join # it check for left table and return matching values from right table if data not found then NAN
```

```python
# Right Join
#right is opposite of left

right_join = pd.merge(df1, df2 , on="ID", how = "right")
right_join
```
