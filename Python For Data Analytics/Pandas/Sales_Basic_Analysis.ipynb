
###Calculating total sales and total profit


```python
total_sales = df['Sales'].sum()
print("The Total Sales is:",total_sales)

total_profit = df['Profit'].sum()
print("The Total Profit is:",total_profit)
```


###Analyzing sales by category


```python
df
```

```python
cat_sales = df.groupby('Category')['Sales'].sum()
print("Sales By Category", cat_sales)
```


###Analyzing profit by category


```python
cat_profit = df.groupby('Category')['Profit'].sum()
print("Profit By Category",cat_profit)
```


###Visualizing sales by category


```python
cat_sales.plot(kind ='bar' , title = 'Sales By Category' , xlabel = "Category" , ylabel ="Sales")
```

###Visualizing profit by category

```python
cat_profit.plot(kind="bar", title="Profit By Category" , xlabel= "Category", ylabel ="Profit")
```
