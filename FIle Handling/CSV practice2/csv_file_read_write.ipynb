##Lets make it dynamic

```python
def adding_data_csv(Name,Age,Location):
  import csv
  with open("csv_read_write.csv","a",newline="") as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow([Name,Age,Location])

Name =input("Enter The Name:")
Age =int(input("Enter The Age :"))
Location =input("Enter The Place Name:")

adding_data_csv(Name,Age,Location)
```

## Working With Dictionary File In Csv


```python
import csv
def create_csv_dict(name ,work,domain):
  keys_data= ["name","work","domain"]
  data =[{"name":name,"work":work,"domain":domain}]
  with open ("employee_data.csv","w",newline="")as file:
    csv_dict_writer=csv.DictWriter(file,fieldnames=keys_data)
    csv_dict_writer.writeheader()
    csv_dict_writer.writerows(data)
name =input("Enter The Name :")
work =input("Enter the Work What you Do?:")
domain=input("Enter Which Domain You work:")

create_csv_dict(name,work,domain)
```
